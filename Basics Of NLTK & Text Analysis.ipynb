```python
txt="Wikipedia is a free enccyclopaedia . Wikipedia is informative . It contains all information . The information are stored in detailed . It is a very useful tool"
```

```python
#printing the text
print(txt)
```

```python
#finding length of the text
len(txt)
```

```python
#converting the text to list which contains sentences as elements
lst=txt.split(".")
lst
```

```python
#converting the text to list which contains words as elements
lst=txt.split(" ")
lst
```

```python
#creating a list of words which have length of words more than 4 
#list comprehension used
lst_new1=[i for i in lst if len(i)>4]
lst_new1
```

```python
#creating a list of words which have length of words less than 4 
#list comprehension used
lst_new2=[i for i in lst if len(i)<4]
lst_new2
```

```python
#creating a list of words that have 1st alphabet of word in capital
#list comprehension used
lst_new3=[i for i in lst if i.istitle()]
lst_new3
```

```python
lst_new4=[i for i in lst if i.endswith("d")]
lst_new4
```

```python
lst_new5=[i for i in lst if i.startswith("i")]
lst_new5
```

```python
#printing all the unique words in the text file
lst_unique=set(lst)
lst_unique=list(lst_unique)
lst_unique
```

```python
#getting all words in lower case
for i in lst:
    print(i.lower())
```

```python
#getting all words in upper case
for i in lst:
    print(i.upper())
```
